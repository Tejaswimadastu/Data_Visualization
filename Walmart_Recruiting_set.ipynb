#Task 2 â€” Data Cleaning

Identify missing values using .isnull().sum().

Fill missing numeric values (Temperature, Fuel_Price, CPI,
Unemployment) using median.

Fill missing markdown-related fields with mean values.

Convert Date column to datetime format.

Remove duplicate rows.

Reset the index after cleaning.

```python
numeric_fill_cols = ["Temperature", "Fuel_Price", "CPI", "Unemployment"]
markdown_cols = ["MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]

print(df.isnull().sum())
print(df.isnull().sum().sum())

for col in numeric_fill_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

for col in markdown_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())

if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"], errors='coerce')

before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print("Duplicates removed:", before - after)

df.reset_index(drop=True, inplace=True)

print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
import matplotlib.pyplot as plt
import numpy as np

cols = [ "Temperature", "Fuel_Price", "CPI"]

for c in cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[c].dropna())
    plt.title(c)
    plt.show()
```

```python
outlier_limits = {}

for c in cols:
    Q1 = df[c].quantile(0.25)
    Q3 = df[c].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outlier_limits[c] = (lower, upper)
outlier_limits
```

```python
df_cap = df.copy()

for c in cols:
    lower, upper = outlier_limits[c]
    df_cap[c] = np.where(df_cap[c] < lower, lower, df_cap[c])
    df_cap[c] = np.where(df_cap[c] > upper, upper, df_cap[c])
```

```python
df_removed = df.copy()

for c in cols:
    lower, upper = outlier_limits[c]
    df_removed = df_removed[(df_removed[c] >= lower) & (df_removed[c] <= upper)]
```
